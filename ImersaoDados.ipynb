**DESAFIO 09**

Ordenar o gráfico countplot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(8,6))

ax=sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'),order = reversed(cod_compostos))
ax.set_title('Top 5 Compostos')
plt.show()
```

**DESAFIO 10**

Melhorar a visualização.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.figure(figsize=(8,6))

ax=sns.countplot(x='composto', data=dados.query('composto in @cod_compostos'), order = reversed(cod_compostos))
ax.set_title('Top 5 Compostos',fontsize=18, fontweight='bold')
plt.xlabel('Compostos',fontsize=15, fontweight='bold')
plt.ylabel('Testes',fontsize=15, fontweight='bold')
```
